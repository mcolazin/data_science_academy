### Multiplicação matrizes


```python
x = torch.Tensor([[2,3], [1,2]])
y = torch.Tensor([[4,5], [5,2]])
x, y
```

```python
# element wise mult
torch.mul(x, y)
```

```python
# division wise
torch.div(x, y)
```

```python
# produto escalar
a, b = torch.Tensor([2,3]), torch.Tensor([4,5])
a, b
```

```python
torch.dot(a,b)
```

```python
### multiplicação de matriz por vetor:
matrix = torch.randn(2,4)
vector = torch.randn(4)
matrix, vector
```

```python
torch.mv(matrix, vector)
```

```python
# mult. matrix e vector e somar um vetor_2 = addmv
vector_2 = torch.randn(2)
torch.addmv(vector_2, matrix, vector)
```

```python
# multiplicando 2 matrizes e adicionar uma matriz => matriz_a + (matriz_x * matriz_y)
torch.addmm(a, x, y)
```

```python
# produto cruzado
c = torch.Tensor([[1], [4], [2]])
d = torch.Tensor([[3], [6], [7]])
print(c, f"\n \n {d}") 
torch.cross(c, d)
```

```python
# MUltiplicação em batch - bmm - muito usado em NLP
batch_1 = torch.randn(10, 3, 4)
batch_2 = torch.randn(10, 4, 5)
batch_mul = torch.bmm(batch_1, batch_2)
print(batch_mul.shape)
print(batch_mul)
```
